# Image to Pencil Sketch with Python

```python
#importing the libraries
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading the image
img = cv2.imread("C://Users//ps505//Documents//Python Scripts//BUTTERFLY.jpg")

```

```python
#Displaying the image
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
#Converting image from BGR to RGB color
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Converting image from BGR to Grey scale
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Inverting the Image
img_invert = cv2.bitwise_not(img_grey)
```

```python
#Blurring the image using GaussianBlur function
img_blur = cv2.GaussianBlur(img_invert, (111,111), 0)
```

```python
#Inverting the blurred image
img_blur_invert = cv2.bitwise_not(img_blur)
```

```python
#Converting the blurred inverted image to sketch by doing bitwise division
img_sketch = cv2.divide(img_grey, img_blur_invert, scale = 256.0)
```

```python
#Plotting the Images
plt.figure(figsize = (18,12))
plt.subplot(1,2,1)
plt.title('Original Image', size = 20)
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Pencil Sketch', size = 20)
sketch_rgb = cv2.cvtColor(img_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(sketch_rgb)
plt.axis('off')

plt.show()
```
